# Fleet analytics

```python
import plotly.graph_objects as go
import pandas as pd

geo = pd.read_csv('CH3_002\gpsTT.csv')

fig = go.Figure(data=go.Scattergeo(
        lon = geo['lon'],
        lat = geo['lat'],
        mode = 'markers',
        ))

fig.update_layout(
        title = 'Self-driving trip',
        geo_scope='usa',
    )
fig.show()
```

```python
ST = pd.read_parquet('CH3_002\subSteering.parquet')
ST.head()
```

```python
import plotly.express as px

fig = px.line(ST, x="time", y="steering_angle", title="Steering Angle")
fig.update_layout(plot_bgcolor="white")
fig.show()
```
